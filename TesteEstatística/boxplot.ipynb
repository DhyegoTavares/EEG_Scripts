```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Dados fornecidos
dados = {
    'Par': ['Frontal_BL01']*7 + ['Frontal_BL02']*7 + ['Frontal_BL03']*7 + ['Frontal_BL04']*7,
    'Diferença': [
        -4.111453, 3.460753, 2.518038, 4.563466, 0.407636, 0.721985, 1.457498,  # Frontal_BL01
        -4.442773, 3.412566, 2.270671, 4.101513, 0.608543, 1.922529, 1.235925,  # Frontal_BL02
        3.445557, 2.727072, 2.456169, 4.610169, 0.163871, 1.201012, 0.389946,  # Frontal_BL03
        1.077673, 3.339079, 1.893324, 2.335918, 0.4925, 0.773199, 1.364539    # Frontal_BL04
    ]
}

df = pd.DataFrame(dados)

# Boxplot
plt.figure(figsize=(10, 6))
sns.boxplot(x='Par', y='Diferença', data=df, palette="husl")
plt.axhline(0, color='gray', linewidth=0.8)
plt.xlabel('Compared Pair')
plt.ylabel('Difference')
plt.title('Distribution of Differences between Pairs')

# Save as PDF
plt.savefig('boxplot_differences.pdf', format='pdf')

plt.show()
```
